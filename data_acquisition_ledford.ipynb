```python
import pandas as pd
```

```python
cherry_dc = pd.read_csv('Q_blooms_dc.csv')
temps_dc = pd.read_csv('Q_temps_dc.csv')
dc_Predtemps_oct = pd.read_csv('dcPredictedTemps_October_1Thru31.csv')
dc_Predtemps_nov = pd.read_csv('dcPredictedTemps_November_1Thru30.csv')
dc_Predtemps_dec = pd.read_csv('dcPredictedTemps_December_1Thru31.csv')
dc_Predtemps_jan = pd.read_csv('dcPredictedTemps_January_1Thru31.csv')
dc_Predtemps_feb = pd.read_csv('dcPredictedTemps_February_20Thru28.csv')
dc_Predtemps_march = pd.read_csv('dcPredictedTemps_March_2026.csv')
dc_Predtemps_april = pd.read_csv("dcPredictedTemps_April_2026.csv")
dc_Predtemps_may = pd.read_csv('dcPredictedTemps_May_2026.csv')
full_dc = pd.concat([cherry_dc, temps_dc,dc_Predtemps_oct, dc_Predtemps_nov, dc_Predtemps_dec, dc_Predtemps_jan, dc_Predtemps_feb, dc_Predtemps_march, dc_Predtemps_april, dc_Predtemps_may])
full_dc.to_csv('dc_cherryAndtemps_full.csv')
print(full_dc.head())
```
